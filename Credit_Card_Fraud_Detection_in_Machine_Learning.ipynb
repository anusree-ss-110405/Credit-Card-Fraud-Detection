```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# loading the dataset to a Pandas DataFrame
h=pd.read_csv('/content/creditcard.csv')
# first 5 rows of the dataset
h.head(5)
```

```python
# dataset informations
h.info()
```

```python
# checking the number of missing values in each column
h.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
h['Class'].value_counts()
```

This data set has fraudulent transactions

0 -> Normal Transaction

1 -> Fraudulent Transaction


```python
# separating the data for analysis
legal_data =h[h.Class==0]
fraud_data =h[h.Class==1]
```

```python
print(legal_data.shape)
print(fraud_data.shape)
```

```python
#statistical measures of data
legal_data.describe()
```

```python
legal_data.Amount.describe()
```

```python
fraud_data.Amount.describe()
```

```python
h.groupby('Class').mean()
```

UnderSampling

```python
# Build a Sample dataset containing similar distribution of normal transaction and fraudulent transaction
# Number of fraudulent transactions is 449
# Perform uniform distribution of data - random sample of 492 transactions
legal_data_sample =legal_data.sample(449)
```

Concatenating two data frames

```python
new_dataset = pd.concat([legal_data_sample,fraud_data],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into features and targets (0 or 1)
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# LogisticRegressionModel - Binary Classification Problems
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation based on accuracy score
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction,Y_test)
```

```python
print('Accuracy score on test data : ',test_data_accuracy)
```
